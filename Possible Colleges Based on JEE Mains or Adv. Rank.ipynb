```python
# importing required libraries

import pandas as pd
import os

PATH=os.getcwd()
print(PATH)
```

```python
# change file name to be worked on
# based on year and round on which you want to find data

data = pd.read_csv("./data/2020-6.csv")
print("Number of Colleges and branch combination in this Year+Round : ", data.shape[0], "\n")

# print( data.info() )


print("5 entries from the top of data: ")
data.head()
```

```python
# Change category
# OPEN/OPEN (PwD)/GEN-EWS/GEN-EWS (PwD)/OBC-NCL/OBC-NCL (PwD)/SC/SC (PwD)/ST/ST (PwD)
df = data[ data['Seat Type'] == "EWS" ]

# Change gender seat choice
# Gender-Neutral/Female-only (including Supernumerary)
df = df[df['Gender']=="Gender-Neutral"]


# for finding IITs only uncomment only 2
# for finding NITs only uncomment only 3
# for finding other than IITs and NITs uncomment 1&4 comment 2&3

df = df[~df.Institute.str.contains("Indian Institute of Technology")]    # by default for colleges based on JEE Mains  
# df = df[df.Institute.str.contains("Indian Institute of Technology")]    
# df = df[df.Institute.str.contains("National Institute of Technology")]  
# df = df[~df.Institute.str.contains("National Institute of Technology")] 


print("Number of Colleges after this filtering: ", df.shape[0], "\n")

# print(df.info())

print("5 entries from the top of data: ")
df.head()
```

```python
# # add line if want to add additional filter
# HS/OS/GO/LA/AI and others if any
# to not include add by using "!=" in seperate lines
# to only see data for that very quota check by using "==", and remove other lines if any

df = df[ df["Quota"]!="HS" ]

print("Number of Colleges after this Filtering: ", df.shape[0])
```

```python
# converting rank from string to int

df['Opening Rank'] = df['Opening Rank'].astype(int)
df['Closing Rank'] = df['Closing Rank'].astype(int)

print("Maximum closing rank scored in this category, gender, HS/OS group as selected above:", end=" ")
print(df["Closing Rank"].max())
```

```python
# change closing rank rank

min_Closing_Rank = 5000
max_Closing_Rank = 10000

possibleCLG = df[df['Closing Rank'] >= min_Closing_Rank ]       
possibleCLG = possibleCLG[possibleCLG['Closing Rank'] <= max_Closing_Rank ]


# if you want to sort the colleges(descending order) based on closing rank uncomment line ending with to sort
# if want to sort in ascending order uncomment ane change "ascending = True"

# possibleCLG.sort_values(by = 'Closing Rank', ascending = False)   # to sort


print("Number of Colleges after this Filtering: ", df.shape[0])

print("5 entries from the top of data: ")
possibleCLG.head()
```

```python
# to see the whole list of data after applying all the filter run this code block
# list will be saved in current working directory
# can be opened with any text editor aur spreadsheet editor

possibleCLG.to_csv("PossibleColleges.csv")
print("File saved to: " + PATH + "\\PossibleColleges.csv")
```
